# Distribución binomial.

$$f(x,n,p)   = {n \choose x} p^x (1-p)^{n-x}$$,  donde: $${n \choose x} = \frac{n!}{x! (n-x)!}$$

Por ejemplo, si tiramos 100 volados, cuya probabilidad es $p=0.5$, entonces $f(20, 100, 0.5)$ nos hablaría de que tan probable es que 20 de esos 100 volados sean favorables.

En otro ejemplo, si tiramos 50 veces un dado, cuyas caras tienen una probabilidad $p=\frac{1}{6}$, entonces $f(10,50,\frac{1}{6})$ nos habla de que tan probable es que 10 de las 50 tiradas serán de un número específico.

```python
import matplotlib.pyplot as plt
import math
```

```python
def binomial(x, n, p):
    factoriales=math.factorial(n)/(math.factorial(x)*math.factorial(n-x))
    probabilidades=p**x * (1-p)**(n-x)
    return factoriales*probabilidades
```

```python
ns=[6, 10, 20, 40, 80, 160]
p=1/6

for n in ns:
    Xs = [k/n     for k in range(0,n+1)]
    Ys=[binomial(x,n,p)      for x in range(0,n+1)]
    plt.plot(Xs, Ys, label=f'n = {n}')

plt.xlabel('Probabilidad $p$')
plt.ylabel('$f(x)$')
plt.legend()
plt.show()
```
