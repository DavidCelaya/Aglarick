# Distribución binomial emergiendo de números aleatorios.
La distribución binomial representa la distribución estadística de una serie de $n$ eventos aleatorios cuando cada uno de los eventos tiene una probabilidad $p$ de ser exitoso. En esta distribución se asume que cada uno de los eventos es no correlacionado. En este ejercicio vamos a graficar la distribución emergiendo espontáneamente de números aleatorios.

---

Aquí importamos la librerías.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Y definimos una función que calcula la cantidad de números que caen en el intervalo $[0, p]$ cuando se lanzan $n$ números al azar en el intervalo $[0, 1]$.

```python
def binomialDis(n,p):
    intentos=[np.random.uniform()    for x in range(0,n)]
    exitos=[intento<=p     for intento in intentos]
    return sum(exitos)
```

Ahora construimos una función que retorna las coordenadas $x$ y $y$ de los puntos que deseamos graficar en una tupla. Esta usa la función anterior para calcular un resultado y acumula los resultados en la variable Ys. De este modo la variable Ys nos dice, de un número de intentos totales dado por puntostotales, cuantas veces aparece el mismo resultado.

```python
def graficaBinomial(puntostotales, n, p):
    Xs=[k/n    for k in range(0,n+1)]
    Ys=[0    for i in range(0,n+1)]
    puntoactual=0
    while puntoactual<puntostotales:
        ubicacion=binomialDis(n, p)
        Ys[ubicacion]+=1
        puntoactual+=1
    return Xs, Ys
```

Y ahora generamos las gráficas resultantes para varios valores de $n$.

```python
ns=[6, 10, 20, 40, 80, 160]
p=1/6
puntos=30000

for n in ns:
    curva=graficaBinomial(puntos, n, p)
    plt.plot(*curva, label=f'n = {n}')

plt.xlabel('Probabilidad $p$')
plt.ylabel('$f(x,n,p)$')
plt.legend()
plt.show()
```

Conforme $n \to \infty$ la curvas tienden hacia un pico progresivamente estrecho centrado en $p$. Conforme $puntos \to \infty$, las curvas se vuelven progresivamente más y más continuas. Podemos lograr cualquier nivel deseado de continuidad en las curvas al incrementar el número los puntos, aunque con la consecuencia de disparar el tiempo de ejecución.
